```python
import numpy as np
import pandas as pd
import arviz as az
import bambi as bmb
import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
from sklearn.metrics import roc_auc_score, matthews_corrcoef
from scipy.stats import ks_2samp
```

```python
dados = pd.read_csv("base_treino.csv")
```

```python
dados.columns
```

```python
dados.rename(columns={j:'VAR'+str(i) for i,j in zip(range(1,11),dados.columns[:-1]) } , inplace=True)
```

Realizando as simulações de inferência para estimar os coeficientes da regressão logística bayesiana.
O amostrador deafult é o No-U-Turn Sampler (NUTS). As prioris são $N(0, 10^{12} I)$

```python
model1 = bmb.Model('''MAU ~ VAR1 + VAR2 + VAR3 + VAR4 + VAR5 + VAR6 + VAR7 + VAR8 + VAR9+ VAR10''', dados, family="bernoulli")
```

```python
fitted1 = model1.fit(draws=2000, chains=2, idata_kwargs={"log_likelihood": True})
```

```python
az.plot_trace(fitted1)
az.summary(fitted1)
```

Notamos pelo primeiro ajuste que as estimativas do intervalo de credibilidade da VAR7, VAR8 e VAR9 contém 0. Assim, essas variáveis deveriam ser retiradas do modelo para realização de um novo ajuste. O novo ajuste sem elas é realizado abaixo. Além disso, o comportamento das simulações das cadeias parecem indicar convergência pois elas se mostram estáveis, dentro de um intervalo fixo e a distribuição marginal assintotica das duas cadeias simuladas são muito similares.

```python
model2 = bmb.Model('''MAU ~ VAR1 + VAR2 + VAR3 + VAR4 + VAR5 + VAR6 + VAR10''', dados, family="bernoulli")
```

```python
fitted2 = model2.fit(draws=2000, chains=2, idata_kwargs={"log_likelihood": True})
```

```python
az.plot_trace(fitted2)
az.summary(fitted2)
```

Aplicando a formula obtida nos dados de teste:

```python
dados_teste = pd.read_csv('base_teste.csv')
```

```python
dados_teste.rename(columns={j:'VAR'+str(i) for i,j in zip(range(1,11),dados_teste.columns[:-1]) } , inplace=True)
```

```python
dados_teste.shape
```

```python
az.summary(fitted2)['mean']
```

```python
coefs = az.summary(fitted2)['mean']
```

Como a forma do predict do objeto model1 é diferente do "padrão" do sklearn, criou-se uma função que retorna o vetor de probabilidades de uma predição dado um conjunto de dados similar ao utilizado no treinamento. Apenas para facilitar as outras tarefas, como o cálculo das métricas de avaliação.

```python
def predict_proba(X , coefs):
    try:
        m1 = X[['VAR1', 'VAR2', 'VAR3', 'VAR4', 'VAR5', 'VAR6','VAR10']]
        m1['intercepto'] = 1
        m1 = m1[['intercepto','VAR1', 'VAR2', 'VAR3', 'VAR4', 'VAR5', 'VAR6','VAR10']]
        return 1/(1+np.exp(-np.dot( m1 , coefs)))
    except:
        return None
```

```python
y_pred = predict_proba(dados_teste , coefs)
```

```python
roc_auc_score( dados_teste['MAU'],  y_pred )
```

```python
matthews_corrcoef( dados_teste['MAU'],  pd.Series(y_pred).apply(lambda x: 1 if x> 0.36 else 0) )
```

```python
ks_2samp( y_pred[ dados_teste['MAU'] == 0 ] , y_pred[ dados_teste['MAU'] == 1 ] )
```
